**Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

**Assign the CSV file to a variable**

```python

# Define the file path
file_path = "/Users/macbook/Downloads/API_SP.POP.TOTL_DS2_en_csv_v2_34/API_SP.POP.TOTL_DS2_en_csv_v2_34.csv"

# Read the CSV file, skipping the metadata rows
df = pd.read_csv(file_path, skiprows=4)
```

**Explore the data**

```python
df.head()
```

```python
df.shape
```

**Isolate Nigeria's population**

```python
Nigeria_df= df[df["Country Name"]== "Nigeria"] 
Nigeria_df
```

**Drop columns that do not contain population data**

```python
values= Nigeria_df.drop(columns=["Country Name", "Country Code", "Indicator Name", "Indicator Code","Unnamed: 69"], axis=1).T
values
```

**Bar Chart**

```python

plt.figure(figsize=(12, 6))  # Increase figure size
plt.bar(values.index, values[174], color='skyblue', edgecolor='black')  # Add color and edge

# Add labels and title
plt.xlabel("Year", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.title("Nigeria Population Over the Years", fontsize=14, fontweight='bold')

# Rotate x-axis labels and limit the number of ticks
plt.xticks(rotation=45, fontsize=10)
plt.locator_params(axis='x', nbins=10)  # Show fewer x-ticks

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the chart
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```
